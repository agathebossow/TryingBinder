```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create synthetic gene expression data
np.random.seed(42)
genes = ['Gene{}'.format(i) for i in range(1, 11)]
samples = ['Sample{}'.format(i) for i in range(1, 6)]
expression_data = np.random.rand(10, 5)

# Create a DataFrame
gene_data = pd.DataFrame(expression_data, index=genes, columns=samples)

# Display the data
gene_data.head()

# Plot gene expression levels
gene_data.plot(kind='bar', figsize=(10, 6))
plt.title('Gene Expression Across Samples')
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.legend(title='Samples')
plt.show()
```
